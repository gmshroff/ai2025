```python
import gym
from stable_baselines3 import A2C
```

```python
env = gym.make("CartPole-v1",max_episode_steps=200,render_mode="human")
```

```python
obs = env.reset()
for i in range(300):
    action=env.action_space.sample()
    obs, reward, terminated, done, info = env.step(action)
    env.render()
    if done or terminated: break
print(i,done)
```

```python
env = gym.make("CartPole-v1",max_episode_steps=200,render_mode=None)
```

```python
model = A2C("MlpPolicy", env, verbose=0)
```

```python
model.learn(total_timesteps=40000)
```

```python
env = gym.make("CartPole-v1",max_episode_steps=200,render_mode="human")
```

```python
obs = env.reset()[0]
for i in range(250):
    action,_ = model.predict(obs, deterministic=True)
    # print(action)
    obs, reward, terminated, done, info = env.step(action)
    env.render()
    if done or terminated: break
print(i,done)
```
